```python
to_replace = ['kg', ' ']

for char in to_replace:
    df1['Weight'] = df1['Weight'].str.replace(char, "", regex=False)

df1['Weight'].head()
```

```python
df2 = pd.read_csv('superhero_powers - superhero_powers.csv')
df2.head()
```

```python
# split string on comma
df2['powers_split'] = df2['Powers'].str.split(',')

# check
df2[['Powers', 'powers_split']].head()
```

```python
exploded = df2.explode('powers_split')
exploded[['hero_names', 'Powers', 'powers_split']].head()
```

```python
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
frames = [df2]

for col in cols_to_make:
    frame = pd.DataFrame()
    frame[col] = df2['Powers'].str.contains(col)
    frames.append(frame)
    merged_powers = pd.concat(frames, axis = 1)
    
# check
merged_powers.head()
```

```python
# drop 'Powers' and 'powers_split' columns
merged_powers.drop(columns = ['Powers', 'powers_split'], inplace = True)

# check
merged_powers.head()
```

```python
# change 'hero_names' column name to match info['Hero']
merged_powers.rename(columns = {'hero_names': 'Hero'}, 
                     inplace = True)

# check
merged_powers.head(2)
```

```python
# check info for merged_powers before merging
merged_powers.info()
```

```python
# check info for info before merging
df1.info()
```

```python
# combine two dataframes on info['Hero'] and merged_powers['Hero']
hero_df = pd.merge(df1, 
                   merged_powers, 
                   on = 'Hero')

# check
hero_df.head()
```

### 1. Compare the average weight of superheroes who have Super Speed to those who do not.

```python
hero_df.groupby('Super Speed')['Weight'].mean().round(2)
```

The average weight of superheroes who have Super Speed is 129.40 kg, and the average weight of those who do not is 101.77 kg.

### 2. What is the average height of heroes for each publisher?

```python
hero_df.groupby('Publisher')['Height'].mean().round(2)
```

In above we have average heights of heros for each publisher in centimeters
